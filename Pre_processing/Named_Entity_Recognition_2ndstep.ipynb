```
def spacy_ner(document):
  return {(ent.text.strip(), ent.label_) for ent in nlp(document).ents}

def NER(dataset):
  ne_list=[]
  for i in range(dataset.shape[0]):

    spcy_ner=spacy_ner(dataset.title[i])
    ne=[]
    for j in spcy_ner:
      
      ne.append(j[0])
    ne_list.append(ne)

  dataset["NE"]=ne_list
  return dataset
```
